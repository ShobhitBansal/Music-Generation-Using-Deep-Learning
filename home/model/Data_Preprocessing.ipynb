```python
p = Path("G:/Github/Music_Generation_Using_Deep_Learning/home/midi_dataset")

notes = []
for file in p.glob("*.mid"):
    print("parsing "+str(file).split("\\")[-1])
    
    midi = converter.parse(file)
    elements = midi.flat.notes
    for e in elements:
        # If a note, add its pitch
        if isinstance(e,note.Note):
            notes.append(str(e.pitch))
            
        if isinstance(e,chord.Chord):
            notes.append("+".join(str(n) for n in e.normalOrder))
```

```python
print("Total Elements : ",len(notes))
print("Unique Elements : ",len(set(notes)))
```

```python
with open("Notes" , 'wb') as f:
    pickle.dump(notes,f)
```
